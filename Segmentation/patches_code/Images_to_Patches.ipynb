```python
import matplotlib.pyplot as plt

def plot_patches(patches):
    num_rows, num_cols, _, _, _, _ = patches.shape
    fig, axes = plt.subplots(num_rows, num_cols, figsize=(8, 8))

    for i in range(num_rows):
        for j in range(num_cols):
            axes[i, j].imshow(patches[i, j, 0])
            axes[i, j].axis('off')
    
    fig.patch.set_facecolor('black')

    plt.subplots_adjust(wspace=0.1, hspace=0.03)
    plt.show()
```

```python
import numpy as np
from matplotlib import pyplot as plt
from patchify import patchify
import cv2

image_file = r"E:\Segmentation\datasets\FloodNet-Supervised_v1.0\train\train-org-img\6414.jpg"
large_image_stack = cv2.imread(image_file)
large_mask_stack = cv2.imread(image_file.replace('org', 'label').replace('jpg', 'png'))

print(large_image_stack.shape)
size = 256
patches_img = patchify(large_image_stack, (size, size, 3), step=size)  
print(patches_img.shape)
```

```python
plt.imshow(large_image_stack)
```

```python
plot_patches(patches_img)
```
